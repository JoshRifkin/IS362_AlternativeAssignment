```python
import pandas as pd
#url = "https://raw.githubusercontent.com/JoshRifkin/IS362_AlternativeAssignment/main/tournamentinfo.txt"
url = "C:/Users/joshz/Documents/CUNY SPS/Fall 2020/Databases 2/Alt Assignment/tournamentinfo.txt"
```

```python
chess = pd.read_csv(url, skiprows=1, sep='|', skip_blank_lines=True)
del chess[' ']
chess = chess.dropna()
even = chess.iloc[::2]
odd = chess.iloc[1::2]
even = even.reset_index()
odd = odd.reset_index()
del even['index']
del odd['index']
```

```python
chess
```

```python
odd
```

```python
names = odd[' Player Name                     ']
names
```

```python
points = odd['Total']
points
```

```python
scores = even[' Player Name                     ']
scores
```

```python
scores = scores.str.split(': ', expand = True) 
scores = scores[1].str.split('->', expand = True)
scores = scores[1:]
scores = scores.reset_index()
del scores['index']
scores
```

```python
state = even.iloc[1::2]
state = state.reset_index()
#dates.columns
state = state[' Pair ']
state
```

```python
players = pd.concat([names, state, points, scores], axis=1, ignore_index=True)
players.columns = ['Player Name', 'State', 'Points', 'Pre-Score', 'Post-Score']
players
```

```python
players.to_csv('chess_players.csv')
```
